```python
import pandas as pd
import numpy as np
import random
import datetime
```

```python
data = pd.read_excel('data.xlsx')
data
```

```python
# add x0 = 1 to every xn
x0_list = [1] * len(data)
data = pd.concat([pd.Series(x0_list), data], axis=1)
data = data.rename(columns={0: 'x0'})
data
```

```python
data_copy = data.copy()
```

```python
y = data_copy['y']
data_copy = data_copy.drop('y', axis=1)
```

```python
def sign(z):
    if z > 0:
        return 1
    else:
        return -1
```

```python
# trial sample
np.array(data_copy.sample(n=1))
```

```python
# trial np.array()
np.array(data_copy.loc[1, :])
```

```python
# trial dot
type(np.dot(w0, np.array(data_copy.loc[1, :])))
```

```python
def pla(data, y):
    w = np.array([0.] * 11)
    N = len(data)
    counter = 1
    while counter <= (5 * N):
        rd_index = random.randint(0, N - 1)
        rd_y = y[rd_index]
        rd_x = np.array(data.iloc[rd_index, :])
        
        # check ans
        if sign(np.dot(w, rd_x)) != rd_y:  # mistake
            counter = 1
            w += rd_y * rd_x
        else:
            counter += 1

    return w
```

```python
a = pla(data_copy, y)
print(a)
np.linalg.norm(a) ** 2
```

```python
# Q13
totalsqrnorm = 0
print('start:', datetime.datetime.now())
for i in range(1000):
    seed = i
    if (seed % 100) == 0:
        print(seed)

    wpla = pla(data=data_copy, y=y)
    totalsqrnorm += np.linalg.norm(wpla) ** 2
print('end:', datetime.datetime.now())
avgsqrnorm = totalsqrnorm / 1000
print(avgsqrnorm)
```

```python
# Q16 unit vector
def get_unit_vector(z):
    unv = z / np.linalg.norm(z)
    return unv
```

```python
temp = np.matrix(data_copy)
for i in range(len(temp)):
    temp[i] = get_unit_vector(temp[i])

temp
```

```python
temp = pd.DataFrame(temp, columns=x_cols)
temp.head(5)
```

```python
totalsqrnorm = 0
print('start:', datetime.datetime.now())
for i in range(1000):
    seed = i
    if (seed % 100) == 0:
        print(seed)

    wpla = pla(data=temp, y=y)
    totalsqrnorm += np.linalg.norm(wpla) ** 2
print('end:', datetime.datetime.now())
avgsqrnorm = totalsqrnorm / 1000
print(avgsqrnorm)
```
