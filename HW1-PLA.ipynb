```python
import pandas as pd
import numpy as np
import random
import datetime
```

```python
data = pd.read_excel('data.xlsx')
data.head(3)
```

```python
# add x0 = 1 to every xn
x0_list = [1.0] * len(data)
data = pd.concat([pd.Series(x0_list), data], axis=1)
data = data.rename(columns={0: 'x0'})
data.head(3)
```

```python
data_copy = data.copy()
```

```python
y = data_copy['y']
data_copy = data_copy.drop('y', axis=1)
```

```python
def sign(z):
    if z > 0:
        return 1
    else:
        return -1
```

```python
data_copy.head(3)
```

```python
# algo
def pla(data, y):
    w = np.array([0.] * 11)
    N = len(data)
    counter = 1
    while counter <= (5 * N):
        rd_index = random.randint(0, N - 1)
        rd_y = y[rd_index]
        rd_x = np.array(data.iloc[rd_index, :])
        
        # check ans
        if sign(np.dot(w, rd_x)) != rd_y:  # mistake
            counter = 1
            w += rd_y * rd_x
        else:
            counter += 1

    return w
```

```python
# try one time
a = pla(data_copy, y)
print(a)
print(np.linalg.norm(a))
np.linalg.norm(a) ** 2
```

```python
# Q13
w_ans = np.array([0.] * 11)
print('start:', datetime.datetime.now())
for i in range(1000):
    wpla = pla(data=data_copy, y=y)
    w_ans += wpla
print('end:', datetime.datetime.now())
w_ans /= 1000
avgsqrnorm = np.linalg.norm(w_ans) ** 2
print(avgsqrnorm)
```

```python
# Q14 Scaling up x by 2
temp = np.matrix(data_copy)
temp *= 2
```

```python
x_cols = list(data_copy.columns)
temp = pd.DataFrame(temp, columns=x_cols)
temp.head(3)
```

```python
# Q15 unit vector
def get_unit_vector(z):
    unv = z / np.linalg.norm(z)
    return unv
```

```python
temp = np.matrix(data_copy)
for i in range(len(temp)):
    temp[i] = get_unit_vector(temp[i])
```

```python
temp = pd.DataFrame(temp, columns=x_cols)
temp.head(3)
```

```python
w_ans = np.array([0.] * 11)
print('start:', datetime.datetime.now())
for i in range(1000):
    wpla = pla(data=temp, y=y)
    w_ans += wpla
print('end:', datetime.datetime.now())
w_ans /= 1000
avgsqrnorm = np.linalg.norm(w_ans) ** 2
print(avgsqrnorm)
```
