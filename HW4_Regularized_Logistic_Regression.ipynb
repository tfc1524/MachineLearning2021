```python
# 16
# preparing train
train_data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/hw4_train.xlsx', header=0)
train_X = train_data.drop('y', axis=1)
train_y = train_data['y']
display(train_data.head(3))
# 3rd order transformation
train_X = np.asarray(train_X)
poly = PolynomialFeatures(3)
train_X = poly.fit_transform(train_X)
train_X = pd.DataFrame(train_X)
print(train_X.shape)
train_X.head(3)

# preparing test
test_data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/hw4_test.xlsx', header=0)
test_X = test_data.drop('y', axis=1)
test_y = test_data['y']
display(test_data.head(3))
# 3rd order transformation
test_X = np.asarray(test_X)
poly = PolynomialFeatures(3)
test_X = poly.fit_transform(test_X)
test_X = pd.DataFrame(test_X)
print(test_X.shape)
test_X.head(3)

avg_acc = 0
for i in range(5):
  # train, select, and predict
  val_y = train_y[40 * i: 40 * (i + 1)]
  val_X = train_X.iloc[40 * i: 40 * (i + 1), :]
  training_y = np.array(train_y.drop(val_y.index, axis=0))
  training_X = np.asarray(train_X.drop(val_X.index, axis=0))
  val_y = np.array(val_y)
  val_X = np.array(val_X)

  print('validating: {}~{}'.format(40 * i, 40 * (i + 1)))
  print('validation size:', len(val_y), len(val_X))
  print('train size:', len(training_y), len(training_X))

  testing_y = np.array(test_y)
  testing_X = np.asarray(test_X)
  prob  = problem(training_y, training_X)
  param = parameter('-s 0 -e 0.000001 -c 5000')
  model = train(prob, param)
  p_labs, p_acc, p_vals = predict(val_y, val_X, model)
  print('Eval:', p_acc)
  print()
  avg_acc += p_acc[0]
avg_acc /= 5
print('Ecv:', avg_acc)
```
