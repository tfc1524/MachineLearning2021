```python
# Q11
train_y_5 = np.array([1. if i == 5 else -1. for i in train_y])
prob  = svm_problem(train_y_5, train_X)
param = svm_parameter('-s 0 -t 0 -c 10')
model = svm_train(prob, param)

coef = model.get_sv_coef()
svs = model.get_SV()
w = np.array([0.] * 36)
for i in range(len(coef)):
  for key in svs[i].keys():
    w[key - 1] += coef[i][0] * svs[i][key]
print(np.linalg.norm(w))
print(w)
```

```python
# Q16
from sklearn.model_selection import train_test_split

gamma_select_count = dict()
gamma_list = [0.1, 1, 10, 100, 1000]
for t in range(1000):
    # ramdomly split data
    train_X_t, val_X_t, train_y_t, val_y_t = train_test_split(train_X, train_y_1, test_size=200)

    min_Eval = np.inf
    min_Eval_gamma = np.inf

    # find best gamma
    for gamma in gamma_list:
        prob  = svm_problem(train_y_t, train_X_t)
        param = svm_parameter('-s 0 -t 2 -c 0.1 -g {}'.format(gamma))
        model = svm_train(prob, param)
        p_label, p_acc, p_val = svm_predict(val_y_t, val_X_t, model)
        Eval = (100 - p_acc[0]) / 100

        if Eval < min_Eval:
            min_Eval = Eval
            min_Eval_gamma = gamma
        elif Eval == min_Eval:
            if gamma < min_Eval_gamma:
                min_Eval = Eval
                min_Eval_gamma = gamma
  
    # count
    if min_Eval_gamma not in gamma_select_count:
        gamma_select_count[min_Eval_gamma] = 1
    else:
        gamma_select_count[min_Eval_gamma] += 1
```
